```python
import pandas as pd

# Đúng cú pháp
df = pd.read_csv('DataCo_UTF8.csv', encoding='utf-8')  # hoặc encoding='ISO-8859-1'
df.head()
```

## Look

```python
df.info()

df.describe()

df.isnull().sum()

#df.dropna(inplace=True)
```

```python
# Liệt kê cột nào là kiểu object
object_columns = df.select_dtypes(include=['object']).columns
print("Các cột kiểu object:", object_columns)
#Liệt kê cột nào là kiểu số
numeric_columns = df.select_dtypes(include=['number']).columns
print("Các cột kiểu số:", numeric_columns)
```

```python
#Liệt kê giá trị unique của mỗi cột kiểu object và in ra tất cả giá trị unique của các cột kiểu object


for col in object_columns:
    unique_values = df[col].unique()
    print(f"Cột '{col}' có {len(unique_values)} giá trị duy nhất:")
    # print(unique_values[:15])  # In ra 15 giá trị đầu tiên
    print(unique_values) # In ra tất cả giá trị unique
    print("\n")

```

```python
# Liệt kê các cột kiểu số và in ra tất cả giá trị unique của các cột kiểu số

for col in numeric_columns:
    unique_values = df[col].unique()
    print(f"Cột '{col}' có {len(unique_values)} giá trị duy nhất:")
    # print(unique_values[:15])  # In ra 15 giá trị đầu tiên
    print(unique_values) # In ra tất cả giá trị unique
    print("\n")

```

```python
#Liệt kê giá trị unique của mỗi cột kiểu object
print("Các giá trị duy nhất của các cột kiểu object:")
for col in object_columns:
    print(f"Cột {col} có {df[col].nunique()} giá trị duy nhất")
    print(df[col].unique()[:10])  # In ra 10 giá trị đầu tiên
    print()
# xuất ra file CSV
df[object_columns].nunique().to_csv('unique_object_columns.csv', header=True)
# Thống kê số lượng giá trị duy nhất của mỗi cột
unique_counts = df.nunique()
print(unique_counts)    
# xuất ra file CSV
unique_counts.to_csv('unique_counts.csv', header=True)
```

```python
#Liệt kê giá trị unique của mỗi cột kiểu số
for col in numeric_columns:
    print(f"Cột {col} có {df[col].nunique()} giá trị duy nhất")
    print(df[col].unique()[:10])  # In ra 10 giá trị đầu tiên
    print()
# xuất ra file CSV
df[numeric_columns].nunique().to_csv('unique_numeric_columns.csv', header=True)
# Thống kê số lượng giá trị duy nhất của mỗi cột
unique_counts_numeric = df[numeric_columns].nunique()
print(unique_counts_numeric)    
# xuất ra file CSV
unique_counts_numeric.to_csv('unique_counts_numeric.csv', header=True)
```
